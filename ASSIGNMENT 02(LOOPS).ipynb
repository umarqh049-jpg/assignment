4. Write a program that takes a number n as input from the user and generates the first 
n terms of the series formed by squaring the natural numbers. 
Sample output 
Enter a number: 6 
The first 6 terms of the series are: 
1 4 9 16 25 36 

```python
# Take input from the user and convert it to an integer
n = int(input("Enter a number: "))

# Generate the series using a list comprehension
# It squares each number from 1 up to n
series = [str(i**2) for i in range(1, n + 1)]

# Join the list elements with commas and print the final output
print(f"The first {n} terms of the series are: {', '.join(series)}")
```

5. Write a program that prompts the user to input a number and prints its multiplication 
table.  

```python
# Prompt the user to input a number
num = int(input("Enter a number to print its multiplication table: "))

print(f"\nMultiplication Table for {num}:")
print("-" * 25)

# Loop 10 times to generate the table from 1 to 10
for i in range(1, 11):
    product = num * i
    print(f"{num} x {i:2d} = {product}")
```

6. Write a program that prompts the user to enter a number n, and then prints all the 
odd numbers between 1 and n.   

```python
# Prompt the user to enter a number
n = int(input("Enter a number: "))

print(f"\nThe odd numbers between 1 and {n} are:")

# Loop from 1 to n (inclusive) with a step of 2
for i in range(1, n + 1, 2):
    # Print numbers on the same line separated by a space
    print(i, end=" ")
print()  # Print a new line at the end
```

7. Write a Python program to print the numbers from 20 to 1 using a while loop. 

```python
# Initialize the counter variable at 20
num = 20

# The loop will run as long as num is greater than or equal to 1
while num >= 1:
    print(num)
    # Decrease the counter by 1 in each iteration
    num -= 1
    
```

8. Write a program that prompts the user to enter a number n and prints all the 
numbers from 1 to n. 

```python
# Prompt the user to enter a number
n = int(input("Enter a number: "))

print(f"\nThe numbers from 1 to {n} are:")

# Loop from 1 to n (inclusive)
for i in range(1, n + 1):
    # Print numbers on the same line separated by a space
    print(i, end=" ")
print()  # Print a new line at the end
```

9. Write a program that prompts the user to enter a number and repeats this process 5 
times. The program should accumulate the numbers entered and then display the final 
running total. 
- Sample Output: 
- Enter a number: 10 
- Enter a number: 15 
- Enter a number: 35 
- Enter a number: 40 
- Enter a number: 50 
- The final running total is: 150   

```python
# Initialize the accumulator variable to 0
running_total = 0

# Loop exactly 5 times
for i in range(5):
    # Prompt the user to enter a number and convert it to an integer
    number = int(input("enter a number: "))
    # Accumulate the number into the running total
    running_total += number

# Display the final running total
print(f"the final running total is : {running_total}")
```

10. Write a Python program that:

1- Creates a list named clean_countries containing the following countries:
[Bangladesh
India
Pakistan
China]

2- Repeatedly asks the user to enter a country name.

3- Checks whether the entered country exists in the clean_countries list.

4-If the country is found in the list:

Display the message:

<country_name> is in clean countries list

Stop the program.

5-If the country is not found, keep asking the user to enter another country.

```python
# 1. Create the list of clean countries
clean_countries = ["Bangladesh", "India", "Pakistan", "China"]

# 2. Start a loop that runs repeatedly
while True:
    # Ask the user to enter a country name
    user_input = input("Enter a country name: ")
    
    # 3. Check whether the entered country exists in the list
    # (Optional: .strip() removes accidental extra spaces, .title() handles capitalization)
    clean_input = user_input.strip().title()
    
    if clean_input in clean_countries:
        # 4. If found, display the success message and stop the program
        print(f"{clean_input} is in clean countries list")
        break
    else:
        # 5. If not found, inform the user and let the loop repeat
        print(f"{user_input} is not in the list. Please try again.\n")
```
